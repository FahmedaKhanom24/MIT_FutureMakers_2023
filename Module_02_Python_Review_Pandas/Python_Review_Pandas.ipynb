# Pandas

Pandas is an open-source Python library that is used for **data manipulation, analysis, and visualization**. It provides data structures for efficiently storing and manipulating large datasets, and includes tools for data cleaning, data transformation, and data analysis.

The two primary data structures provided by Pandas are Series and DataFrame.

Pandas is widely used in data analysis and data science, as it provides a convenient and efficient way to work with **large datasets**. It is also often used in conjunction with other libraries, such as NumPy, Matplotlib, and Scikit-learn, to perform advanced data analysis and machine learning tasks.

## Series

A Series is a **one-dimensional labeled array** that can hold data of any type (integer, float, string, etc.). A Series is similar to a column in a spreadsheet or a SQL table.

To create a Series in Pandas, you can use the `pd.Series()` function. Here's an example

```python
# Create a Series of integers
my_series = pd.Series([10, 20, 30, 40, 50])

print('A Series of integers:')
print(my_series)
print("\n")

# Create a Series of strings with custom index labels
my_series = pd.Series(['apple', 'banana', 'cherry', 'date', 'elderberry'], index=['a', 'b', 'c', 'd', 'e'])

print('A Series of strings:')
print(my_series)
```

## Dataframes

A DataFrame is a **two-dimensional table** that can hold data of **multiple** types, similar to a spreadsheet or a SQL table.

A DataFrame consists of rows and columns, where each column can have a different data type (e.g., integer, float, string, boolean, datetime).

```python
# Create a DataFrame from a dictionary of lists
data = {'name': ['Alice', 'Bob', 'Charlie', 'David', 'Emily'],
        'age': [25, 30, 35, 40, 45],
        'city': ['New York', 'Paris', 'London', 'Tokyo', 'Sydney']}

df = pd.DataFrame(data)

print('A DataFrame created from a dictionary where keys are strings and values are lists of strings or ints:')
print(df)
```

```python
# Create a DataFrame with custom row and column labels
data = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
df = pd.DataFrame(data, index=['row1', 'row2', 'row3'], columns=['col1', 'col2', 'col3'])

print('A DataFrame created from a 2D array with custom headers of rows and columns')
print(df)
```

# Dataframe Metadata

Pandas includes a wide range of built-in functions for working with these data structures, including grouping, aggregating, merging, and reshaping data.

```python
data = {'name': ['Alice', 'Bob', 'Charlie', 'David', 'Emily'],
        'age': [25, 30, 35, 40, 45],
        'city': ['New York', 'Paris', 'London', 'Tokyo', 'Sydney']}

df = pd.DataFrame(data)

print('Number of rows and columns in the DataFrame:')
print(df.shape)
print("-"*100)

print('Description of Indices:')
print(df.index)
print("-"*100)

print('Description of Columns:')
print(df.columns)
print("-"*100)

print('Information about the DataFrame:')
print(df.info())
print("-"*100)

print("Number of non-NA values")
print(df.count())
print("-"*100)

print("Descriptive statistics for the numerical columns")
print(df.describe())
```
